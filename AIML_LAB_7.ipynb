```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

**GET THE DATA**

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer=load_breast_cancer()
```

```python
cancer.keys()
```

**DESCRIPTION OF FEATURES ARE AS FOLLOWS:**

```python
print(cancer['DESCR'])
```

```python
cancer['feature_names']
```

**SET THE DATAFRAME**

```python
df=pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
df.info()
```

```python
df.describe()
```

```python
np.sum(pd.isnull(df).sum())
```

**TARGET DATA IN DATASET**

```python
cancer['target']
```

```python
df['Cancer']=pd.DataFrame(cancer['target'])
df.head()
```

**EXPLORATORY DATA ANALYSIS**

```python
sns.set_style('whitegrid')
sns.countplot(x='Cancer',data=df,palette='RdBu')
```

```python
l=list(df.columns[0:10])
for i in range(len(l)-1):
  sns.boxplot(x='Cancer',y=l[i],data=df,palette='rainbow')
  plt.figure()
```

**NOT ALL THE FEATURES:**

```python
f,(ax1,ax2)=plt.subplots(1,2,sharey=True,figsize=(12,6))
ax1.scatter(df['mean area'],df['Cancer'])
ax1.set_title('Cancer cases as a function of mean area',fontsize=15)
ax2.scatter(df['mean smoothness'],df['Cancer'])
ax2.set_title('Cancer cases as a function of mean smoothness',fontsize=15)
```
